```python
# Polynomial regression analysis

%matplotlib inline
from pylab import *
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

# create seed value -> run the code return same result for consistancy.(not all over the place)
np.random.seed(2)

pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds

# Display original scatterplot before apply regression
scatter(pageSpeeds, purchaseAmount)
```

```python
# Try 3rd degree polynomial:
# prep for polynomial fit
x = np.array(pageSpeeds)
y = np.array(purchaseAmount)

# 3rd degree polynomial 
p4 = np.poly1d(np.polyfit(x, y, 3))

# start from 0-7 
xp = np.linspace(0, 7, 100)
plt.scatter(x, y)
plt.plot(xp, p4(xp), c='r')
plt.show()
# overfitting of outliers. Not good fit
```

```python
# 4th degree polynomial
# prep for polynomial fit
x = np.array(pageSpeeds)
y = np.array(purchaseAmount)

p4 = np.poly1d(np.polyfit(x, y, 4))

# start from 0-7 
xp = np.linspace(0, 7, 100)
plt.scatter(x, y)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
# get r2 /r_squared
r2 = r2_score(y, p4(x))

print(r2)
```
