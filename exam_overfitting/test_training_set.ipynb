```python
# apply polynomial regression on test&training datasets to exam overfitting
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
from sklearn.metrics import r2_score

# set seed value
np.random.seed(2)

# generate normalized random data 
pageSpeeds = np.random.normal(2.5, 1.0, 100)
purchaseAmount = np.random.normal(35.0, 15.0, 100) / pageSpeeds

# generate scatterplot -> exponential
scatter(pageSpeeds, purchaseAmount)
```

```python
# spliting dataset
# take 80% of the data for training
trainX = pageSpeeds[:80]
# rest 20% for the test
testX = pageSpeeds[80:]

trainY = purchaseAmount[:80]
testY = purchaseAmount[80:]
```

```python
# viz of training data
scatter(trainX, trainY)
```

```python
# viz of testing data
scatter(testX, testY)
```

```python
# polynomial fitting 
# transition into array
x = np.array(trainX)
y = np.array(trainY)

p4 = np.poly1d(np.polyfit(x, y, 7))
```

```python
# visualize the regression

# set range 0-7 and 100 data points
xp = np.linspace(0, 7, 100)

# create axes
axes = plt.axes()

# x,y axes limits
axes.set_xlim([0,7])
axes.set_ylim([0, 200])

# plotting against training data
plt.scatter(x, y)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
# plot polynomial against testing data
testx = np.array(testX)
testy = np.array(testY)

axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0, 200])
plt.scatter(testx, testy)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
# Evaluate the r2-score on training set
r2 = r2_score(testy, p4(testx))
print(r2)
```

```python
# Evaluate the r2-score on testing set
r2 = r2_score(np.array(trainY), p4(np.array(trainX)))
print(r2)
```

```python
# in conclusion -> overfitting
# overfitting works well in training set and bad in testing set
```
