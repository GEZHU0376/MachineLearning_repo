```python
import sys
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# generate 2 d mesh grid
nx, ny = (100, 100)

# unit
x = np.linspace(0,10,nx)
y = np.linspace(0,10,ny)

xv, yv = np.meshgrid(x,y)
```

```python
# define a function to plot
def f(x,y):
    return x * (y**2)

# calculate z value for each x,y plot
z = f(xv, yv)
```

```python
# color plot in matplotlib
# display the data first then calculate the gradients
plt.figure(figsize=(14,12))
plt.pcolor(xv,yv,z)
plt.title("2D plot of f(x,y)=xy^2")
plt.colorbar()
plt.show()
```

```python
# generate a 2d meshgrid for the gradient
nx, ny = (10,10)
x = np.linspace(0,10,nx)
y = np.linspace(0,10,ny)
xg,yg = np.meshgrid(x,y)

# calculate the gradient of f(x,y)
# numpy dealing with row first so we need to move y in front of x
Gy, Gx = np.gradient(f(xg,yg))

plt.figure(figsize=(14,12))
plt.pcolor(xv,yv,z)
plt.title("Gradient plot of f(x,y)=xy^2")
plt.colorbar()
plt.quiver(xg, yg, Gx, Gy, scale = 1000, color = 'w')
plt.show()
```

```python
# calculate the gradient of f(x,y) = xy^2
def ddx(x,y):
    return y**2

def ddy(x,y):
    return 2*y*x

Gx = ddx(xg, yg)
Gy = ddy(xg, yg)
```
