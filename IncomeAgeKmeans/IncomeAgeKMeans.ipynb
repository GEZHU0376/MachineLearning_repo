Income vs Age K-mean clustering 
Data: Kaggle toy dataset
https://www.kaggle.com/carlolepelaars/toy-dataset


```python
from numpy import random, array
%matplotlib inline
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from numpy import random, float

#Create fake income/age clusters for N people in k clusters
def createClusteredData(N, k):
    # consist random seed value set
    random.seed(10)
    pointsPerCluster = float(N)/k
    X = []
    for i in range (k):
        # create random centroid of income and age
        # using the toy dataset from kaggle
        incomeCentroid = random.uniform(647.0, 177175.0)
        ageCentroid = random.uniform(25.0, 65.0)
        for j in range(int(pointsPerCluster)):
            X.append([random.normal(incomeCentroid, 10000.0), random.normal(ageCentroid, 2.0)])
    X = array(X)
    return X

# create 100 people with four clusters
data = createClusteredData(100, 4)

# create k mean model
model = KMeans(n_clusters=4)

# Note I'm scaling the data to normalize it! Important for good results.
model = model.fit(scale(data))

# We can look at the clusters each data point was assigned to
print(model.labels_)

# And we'll visualize it:
plt.figure(figsize=(8, 6))
plt.scatter(data[:,0], data[:,1], c=model.labels_.astype(float))
plt.show()
```
